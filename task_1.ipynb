```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the dataset is in a CSV file, you can read it into a pandas DataFrame
data = pd.read_csv('Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_5871594.csv')

# Group the data by 'Region' and 'IncomeGroup' and count the occurrences
grouped_data = data.groupby(['Region', 'IncomeGroup']).size().unstack()

# Plotting the bar chart
grouped_data.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Income Group Count by Continent', fontsize=16)
plt.xlabel('Continent', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Income Group')
plt.show()
```
